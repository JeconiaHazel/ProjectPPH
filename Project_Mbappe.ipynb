## Data cleaning

### Subtask:
Clean the data by handling missing values and correcting data type inconsistencies.


**Reasoning**:
Clean the data by handling missing values, correcting data type inconsistencies, and removing duplicates.



```python
# Handle missing values in numerical columns using the median
numerical_cols = ['laba_bersih_normal', 'laba_sebelum_pajak', 'pendapatan_operasional']
for col in numerical_cols:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)

# Handle missing values in other columns
# Remove rows with missing values for columns with a small number of missing values
threshold = 5  # Example threshold, adjust as needed
for col in df.columns:
    if df[col].isnull().sum() > 0 and df[col].isnull().sum() <= threshold and col not in numerical_cols:
        df.dropna(subset=[col], inplace=True)
    elif df[col].isnull().sum() > threshold and col not in numerical_cols:
        # Impute with 'Missing' for columns with many missing values
        df[col].fillna('Missing', inplace=True)

# Correct data type inconsistencies
for col in ['persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan',
            'total_hpp', 'pendapatan_penjualan', 'ppn_keluaran', 'ppn_masukan', 'ppn_netto',
            'laba_kotor_penjualan', 'beban_operasional', 'overhead_farmasi', 'penyusutan_total',
            'penyusutan_garis_lurus', 'penyusutan_saldo_menurun', 'penyusutan_fiskal',
            'laba_kotor_operasional', 'laba_kena_pajak', 'pph_badan_normal',
            'pph_badan_tax_holiday', 'laba_bersih_tax_holiday', 'kerugian_fiskal']:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace('.', '').str.replace(',', '.'), errors='coerce')


# Remove duplicate rows
num_duplicates = df.duplicated().sum()
if num_duplicates > 0:
    print(f"Removed {num_duplicates} duplicate rows.")
    df.drop_duplicates(inplace=True)

# Print the shape of the DataFrame
print(f"DataFrame Shape after cleaning: {df.shape}")

display(df.head())
```

## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Perform data wrangling and feature engineering as instructed.



```python
# Convert relevant columns to numeric, handling errors
for col in ['persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan',
            'total_hpp', 'pendapatan_operasional', 'pendapatan_penjualan', 'ppn_keluaran', 'ppn_masukan', 'ppn_netto',
            'laba_kotor_penjualan', 'beban_operasional', 'overhead_farmasi', 'penyusutan_total',
            'penyusutan_garis_lurus', 'penyusutan_saldo_menurun', 'penyusutan_fiskal',
            'laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_kena_pajak', 'pph_badan_normal',
            'laba_bersih_normal', 'laba_bersih_tax_holiday', 'kerugian_fiskal']:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace('.', '').str.replace(',', '.'), errors='coerce')

# Feature Engineering
df['profit_margin'] = df['laba_bersih_normal'] / df['pendapatan_operasional']
df['tax_burden'] = df['pph_badan_normal'] / df['laba_kena_pajak']
df['operating_expense_ratio'] = df['beban_operasional'] / df['pendapatan_operasional']

# Inventory Turnover (assuming 'persediaan_awal' and 'persediaan_akhir' are available)
df['avg_inventory'] = (df['persediaan_awal'] + df['persediaan_akhir']) / 2
df['inventory_turnover'] = df['total_hpp'] / df['avg_inventory']

# Handle potential infinite or NaN values after calculations
df.replace([float('inf'), -float('inf')], float('nan'), inplace=True)
df.fillna(0, inplace=True)

# Display the updated DataFrame
display(df.head())
```

## Data analysis

### Subtask:
Analyze the key financial metrics and relationships within the data.


**Reasoning**:
Calculate descriptive statistics, explore correlations, analyze trends, and investigate outliers for key financial metrics.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Descriptive Statistics
descriptive_stats = df[['tahun', 'pendapatan_operasional', 'laba_bersih_normal', 'profit_margin', 'tax_burden', 'operating_expense_ratio', 'inventory_turnover']].groupby('tahun').describe()
display(descriptive_stats)

# 2. Correlation Matrix and Heatmap
correlation_matrix = df[['profit_margin', 'tax_burden', 'operating_expense_ratio', 'inventory_turnover', 'pendapatan_operasional', 'laba_bersih_normal']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Financial Metrics')
plt.show()

# 3. Trend Analysis
plt.figure(figsize=(12, 6))
sns.lineplot(x='tahun', y='pendapatan_operasional', data=df, label='Revenue', marker='o', color='blue')
sns.lineplot(x='tahun', y='laba_bersih_normal', data=df, label='Net Profit', marker='o', color='green')
sns.lineplot(x='tahun', y='beban_operasional', data=df, label='Operating Expenses', marker='o', color='red')
plt.title('Trend Analysis of Key Financial Metrics Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()

# 4. Outlier Analysis
plt.figure(figsize=(12, 6))
metrics = ['pendapatan_operasional', 'laba_bersih_normal', 'profit_margin', 'operating_expense_ratio']
for i, metric in enumerate(metrics):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[metric])
    plt.title(f'Box Plot of {metric}')
plt.tight_layout()
plt.show()
```
