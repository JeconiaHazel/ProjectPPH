## Data cleaning

### Subtask:
Clean the data by handling missing values, outliers, and data type conversions. Remove duplicate rows.


**Reasoning**:
Handle missing values, outliers, and data type conversions. Remove duplicate rows.



```python
# Fill missing values in numerical columns with the median
numerical_cols = ['Tax_Holiday_Akhir', 'Pendapatan', 'Beban_Operasional', 'Penyusutan', 'Nilai_Perolehan', 'Umur_Ekonomis']
for col in numerical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# Fill missing values in categorical columns with the mode
categorical_cols = ['Skenario', 'Aset_ID', 'Kategori', 'Metode']
for col in categorical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])

# Remove duplicate rows
df.drop_duplicates(inplace=True)
print(f"Number of duplicate rows removed: {df.duplicated().sum()}")

# Convert data types if necessary (example)
# If 'Tahun' should be a datetime, but is not
# df['Tahun'] = pd.to_datetime(df['Tahun'], format='%Y')

# Outlier handling using IQR method for numerical columns
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

display(df.head())
```
